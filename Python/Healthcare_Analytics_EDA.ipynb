Univariate Analysis

```python
#Patient Distribution by Age Group
age = df['age_group'].value_counts().sort_index()

plt.figure(figsize=(7,5))
plt.bar(age.index, age.values)

plt.title("Patient Distribution by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Patients")

plt.show()
```

 Shows the patient distribution across different age groups.
 Helps hospital management identify the primary patient population.

```python
#Gender Distribution
gender = df['gender'].value_counts()

plt.figure(figsize=(5,5))
plt.pie(
    gender.values,
    labels=gender.index,
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Gender Distribution")

plt.show()
```

Compares male and female patient visits.
Useful for demographic planning.

```python
#Region Distribution
region = df['region'].value_counts()

plt.figure(figsize=(8,5))
plt.bar(region.index, region.values)

plt.title("Patients by Region")
plt.xlabel("Region")
plt.ylabel("Patients")

plt.xticks(rotation=20)

plt.show()
```

Identifies regions with the highest patient demand.
 Helps hospital expansion and outreach planning.

```python
#Department Distribution
dept = df['department'].value_counts()

plt.figure(figsize=(8,5))
plt.bar(dept.index, dept.values)

plt.title("Patients by Department")
plt.xlabel("Department")
plt.ylabel("Patients")

plt.xticks(rotation=20)

plt.show()
```

 Departments with higher patient counts may require additional staffing and resources.

```python
#Visit Type Distribution
visit = df['visit_type'].value_counts()

plt.figure(figsize=(5,5))

plt.pie(
    visit.values,
    labels=visit.index,
    autopct='%1.1f%%'
)

plt.title("Visit Type Distribution")

plt.show()
```

Shows the proportion of emergency and routine visits.
Useful for workforce and emergency preparedness planning.

```python
#Treatment Type Distribution
treat = df['treatment_type'].value_counts()

plt.figure(figsize=(8,5))

plt.bar(treat.index, treat.values)

plt.title("Treatment Type Distribution")
plt.xlabel("Treatment Type")
plt.ylabel("Patients")

plt.xticks(rotation=30)

plt.show()
```


 Identifies the most frequently delivered treatments.
Supports equipment and inventory planning.

Correlation Analysis

```python
#Select Numerical Columns
numerical_columns = [
    'length_of_stay_days',
    'treatment_cost',
    'recovery_score',
    'readmission_risk'
]

df[numerical_columns].head()
```

```python
#Correlation Matrix
correlation = df[numerical_columns].corr()

correlation
```

```python
#Heatmap
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))

sns.heatmap(
    correlation,
    annot=True,
    cmap='Blues',
    linewidths=0.5
)

plt.title("Correlation Between Hospital Metrics")
plt.show()
```

The correlation matrix helps identify relationships between treatment cost, hospital stay duration, recovery score, and readmission risk. Strong correlations can help management understand which operational factors influence patient outcomes and healthcare costs.

```python
#Treatment Cost vs Length of Stay
plt.figure(figsize=(8,6))

plt.scatter(
    df['length_of_stay_days'],
    df['treatment_cost']
)

plt.xlabel("Length of Stay (Days)")
plt.ylabel("Treatment Cost")
plt.title("Treatment Cost vs Length of Stay")

plt.show()
```

Longer hospital stays generally tend to increase treatment costs due to extended medical care and resource utilization.

```python
#Recovery Score vs Readmission Risk
plt.figure(figsize=(8,6))

plt.scatter(
    df['recovery_score'],
    df['readmission_risk']
)

plt.xlabel("Recovery Score")
plt.ylabel("Readmission Risk")

plt.title("Recovery Score vs Readmission Risk")

plt.show()
```

This visualization helps assess whether improved recovery outcomes are associated with lower chances of patient readmission.

Business Recommendations

4. Revenue Planning

Use monthly revenue trends to improve budgeting, resource planning, and future financial forecasting.

5. Focus on Patient Recovery
